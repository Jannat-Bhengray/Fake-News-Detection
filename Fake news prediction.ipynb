```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score
from sklearn.model_selection import train_test_split
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import numpy as np
```

```python
import nltk
nltk.download('stopwords')
```

```python
df=pd.read_csv(r'C:\Users\vikas\Documents\My Projects dataset\archive.zip')
```

```python
df['content']= df['title']+ ' ' + df['source_domain']
```

```python
x=df[['title','news_url','source_domain','tweet_num','content']]
y=df['real']
```

```python
print(x)
print(y)
```

```python
port_stem=PorterStemmer()
```

```python
def stemming(content):
    stemmed_content= re.sub('[^a-zA-Z]', ' ',content)
    stemmed_content= stemmed_content.lower()
    stemmed_content= stemmed_content.split()
    stemmed_content= [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content= ' '.join(stemmed_content)
    return stemmed_content
```

```python
df['content']= df['content'].apply(stemming)
```

```python
print(df['content'])
```

```python
x=df['content'].values
y=df['real'].values

print(x)
```

```python
print(y)
```

```python
# to transform textual data in numeric value
```

```python
vectorize= TfidfVectorizer()
vectorize.fit(x)
x= vectorize.transform(x) 
```

```python
print(x)
```

```python
x=x.toarray()   # to conver spars array in normal array
print(x)
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.2, random_state=2)
```

```python
# trainng model on logistic regression  
```

```python
logic_r= LogisticRegression()
logic_r.fit(x_train,y_train)
y_pred= logic_r.predict(x_test)
```

```python
print(y_pred)
```

```python
#checking accuracy
```

```python
accuracy=accuracy_score(y_pred,y_test)
print('accuracy is',accuracy)
```

```python
x_train_pred=logic_r.predict(x_train)
accuracy_t=accuracy_score(x_train_pred,y_train) 
print('accuracy is:',accuracy)
```
